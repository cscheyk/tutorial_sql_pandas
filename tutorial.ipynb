# Load File

```python
import pandas as pd
import sqlite3

data = pd.read_excel('data/Sample - Superstore.xls')
data.head()
```

# Publish data in sqlite

```python
conn = sqlite3.connect('example_database')
cursor = conn.cursor()

cursor.execute('''
    CREATE TABLE IF NOT EXISTS orders (
        RowID number,
        OrderID text,
        OrderDate date,
        ShipDate date,
        ShipMode text,
        CustomerID text,
        CustomerName text,
        Segment text,
        Country text,
        City text,
        State text,
        PostalCode number,
        Region text,
        ProductID text,
        Category text,
        SubCategory text,
        ProductName text,
        Sales number,
        Quantity number,
        Discount number,
        Profit number)''')
conn.commit()
```

```python
n_rows = data.to_sql('orders', conn, if_exists='replace', index = False)
```

# Example 1: Count rows

```python
sql = 'select count(*) from orders'

pd.read_sql(sql, conn)
```
